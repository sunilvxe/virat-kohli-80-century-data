```python
#1.Check for missing values
df.isnull().sum()
```

```python
# Convert 'Runs' and 'S/R' to numeric (if necessary)
df['Runs'] = pd.to_numeric(df['Runs'], errors='coerce')
df['S/R'] = pd.to_numeric(df['S/R'], errors='coerce')

# Numerical Columns for Analysis
numerical_cols = ['No.', 'Runs', 'Pos.', 'Inn.', 'S/R']

# Summary Statistics for Numerical Columns
print(df[numerical_cols].describe())

# Plot histograms for numerical columns
for col in numerical_cols:
    plt.figure(figsize=(8, 5))
    sns.histplot(df[col], bins=30, kde=True)
    plt.title(f'Histogram of {col}')
    plt.xlabel(col)
    plt.ylabel('Frequency')
    plt.show()

# Boxplots for Numerical Columns to check for outliers (Horizontal orientation)
# Ensure 'Against' is categorical and doesn't have any missing values
if df['Against'].dtype == 'object':
    df['Against'] = df['Against'].astype('category')

for col in numerical_cols:
    plt.figure(figsize=(8, 5))
    sns.boxplot(x=df[col], y=df['Against'])  # 'x' should be numeric, 'y' should be categorical
    plt.title(f'Boxplot of {col} (Horizontal orientation)')
    plt.show()
```

```python
# Categorical Columns for Analysis
categorical_cols = ['Against', 'Pos.', 'Test', 'Venue', 'H/A/N', 'Result', 'Match Type']

# 7. Frequency counts for Categorical Columns
for col in categorical_cols:
    print(f"Value counts for {col}:\n", df[col].value_counts())
    print("\n" + "-"*50)

# Bar Plots for Categorical Data (Vertical orientation)
for col in categorical_cols:
    plt.figure(figsize=(15, 8))
    sns.countplot(data=df, x=col, palette="Set2")  # Correct placement for vertical barplot
    plt.title(f'Bar Plot of {col}')
    plt.xlabel(col)
    plt.ylabel('Count')
    plt.xticks(rotation=45)
    plt.show()
```

# **bivariate analysis**

```python
df.columns
```

```python
for i in range(len(numerical_cols)):
    for j in range(i + 1, len(numerical_cols)):
        plt.figure(figsize=(8, 5))
        sns.scatterplot(data=df, x=numerical_cols[i], y=numerical_cols[j])
        plt.title(f'Scatter plot between {numerical_cols[i]} and {numerical_cols[j]}')
        plt.show()
```

```python
# Violin plots for Categorical vs Numerical (e.g., 'Against' vs 'Runs')
for col in numerical_cols:
    plt.figure(figsize=(8, 5))
    sns.violinplot(x=df['Against'], y=df[col])
    plt.title(f'Violin Plot of {col} by Against')
    plt.xlabel('Against')
    plt.ylabel(col)
    plt.xticks(rotation=45)
    plt.show()
```

```python
# Count plot for two categorical variables (e.g., 'Against' vs 'Match Type')
plt.figure(figsize=(15, 8))
sns.countplot(data=df, x='Against', hue='Match Type')
plt.title('Count Plot of Against vs Match Type')
plt.xlabel('Against')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

```python
colors = {
    'Draw': 'lightcoral',       # Light Red for 'Draw'
    'Lost': 'gray',             # Gray for 'Lost'
    'Won': 'darkblue',          # Dark Blue for 'Won'
    'Lost D/L': 'orange',       # Orange for 'Lost D/L'
    'Tied': 'yellow',           # Yellow for 'Tied'
    'Won D/L': 'green'          # Green for 'Won D/L'
}

# Plot stacked bar plot
# Apply the colors to the categories based on the crosstab's columns
crosstab = pd.crosstab(df['Against'], df['Result'])
crosstab.plot(kind='bar', stacked=True, figsize=(15, 8), color=[colors.get(col, 'lightgray') for col in crosstab.columns])

# Set plot title and labels
plt.title('Stacked Bar Plot of Against vs Result')
plt.xlabel('Against')
plt.ylabel('Count')

# Rotate x-axis labels for better readability
plt.xticks(rotation=45)

# Display the plot
plt.show()
```

```python
# Convert 'Runs' to numeric, assuming that centuries are above 99 runs
df['Runs'] = pd.to_numeric(df['Runs'], errors='coerce')

# Filter the DataFrame to only include rows where 'Runs' >= 100 (centuries)
centuries_df = df[df['Runs'] >= 100]

# Group by opponent ('Against') and count the number of centuries
centuries_by_opponent = centuries_df['Against'].value_counts()

# Plot the total centuries by opponent
plt.figure(figsize=(15, 8))
centuries_by_opponent.plot(kind='bar', color='royalblue')
plt.title('Total Centuries by Opponent')
plt.xlabel('Opponent')
plt.ylabel('Number of Centuries')
plt.xticks(rotation=45)
plt.show()
```

```python
# Drop rows where either 'Runs' or 'S/R' is NaN
df_cleaned = df.dropna(subset=['Runs', 'S/R'])

# Create a figure and axis for the plot
fig, ax1 = plt.subplots(figsize=(15, 8))

# Plot histogram for 'Runs' on the first axis (ax1)
sns.histplot(df_cleaned['Runs'], bins=30, kde=False, color='skyblue', ax=ax1)
ax1.set_xlabel('Runs (Score)', fontsize=12)
ax1.set_ylabel('Frequency', fontsize=12)
ax1.set_title('Histogram of Runs and Line Plot for Strike Rate', fontsize=14)

# Create a second axis for the line plot (ax2)
ax2 = ax1.twinx()

# Plot line plot for 'Strike Rate (S/R)' vs 'Runs' on the second axis (ax2)
sns.lineplot(x=df_cleaned['Runs'], y=df_cleaned['S/R'], color='darkblue', ax=ax2)
ax2.set_ylabel('Strike Rate (S/R)', fontsize=12)

# Display the plot
plt.show()
```

```python
df['Runs'] = pd.to_numeric(df['Runs'], errors='coerce')

# Filter for rows where the player has scored a century (Runs >= 100)
centuries_df = df[df['Runs'] >= 100]

# Count the number of centuries by venue
centuries_by_venue = centuries_df['Venue'].value_counts()

# Plot the results as a bar chart
plt.figure(figsize=(15, 8))
centuries_by_venue.plot(kind='bar', color='skyblue')
plt.title('Centuries by Venue', fontsize=14)
plt.xlabel('Venue', fontsize=12)
plt.ylabel('Number of Centuries', fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
centuries_by_position = centuries_df['Pos.'].value_counts()

# Plot the results as a bar chart
plt.figure(figsize=(10, 6))
centuries_by_position.plot(kind='bar', color='lightgreen')
plt.title('Centuries by Batting Position', fontsize=14)
plt.xlabel('Batting Position', fontsize=12)
plt.ylabel('Number of Centuries', fontsize=12)
plt.xticks(rotation=0)  # Keeps the batting positions horizontally aligned
plt.tight_layout()
plt.show()
```

```python
# Count the number of centuries by innings
centuries_by_innings = centuries_df['Inn.'].value_counts()

# Plot the results as a bar chart
plt.figure(figsize=(10, 6))
centuries_by_innings.plot(kind='bar', color=['lightblue', 'salmon'])

# Add title and labels
plt.title('Centuries by Innings', fontsize=14)
plt.xlabel('Innings', fontsize=12)
plt.ylabel('Number of Centuries', fontsize=12)
plt.xticks(rotation=0)  # Keeps the x-axis labels horizontally aligned

# Display the plot
plt.tight_layout()
plt.show()
```
